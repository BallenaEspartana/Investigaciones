#Caso de Estudio

Para efectos de estudio, se analizaran los datos de los titulados cuya casa de estudio este en Concepción

```python
#Para asegurarnos de tomar todas las variaciones posibles de concepcion quitamos minisculas y tildes
df['comuna_sede_limpia'] = df['comuna_sede'].apply(lambda x: unidecode.unidecode(str(x)).upper())
dfConcepcion = df[df['comuna_sede_limpia'] == 'CONCEPCION']
```

```python
dfConcepcion.head(5)
```

```python
dfConcepcion.shape
```

**Interpretación**
+ Hay 20068 Titulados de Concepcion desde 2007 hasta 2023

```python
tituladosTotal=len(df)
tituladosConce=len(dfConcepcion)
tituladosNoConce=tituladosTotal-tituladosConce

plt.figure(figsize=(10, 6))
plt.pie(
    [tituladosConce,tituladosNoConce],
    labels=['Titulados de Concepción','Titulados Fuera de Concepcion'],
    autopct='%1.1f%%',
    colors=['#66b3ff', '#ff9999'],
    startangle=90,
    )
plt.title('Distribución de titulados de educacion superior por comuna: Concepción vs Otras (2007 a 2023)')
plt.axis('equal')
plt.show()
```
